```python
def recommend_gv_tracks(gv_name):
    gv_tracks_id = np.where(gv_pivot.index == gv_name)[0][0]
    distance, suggestion = model.kneighbors(gv_pivot.iloc[gv_tracks_id,:].values.reshape(1,-1), n_neighbors=6 )
    
    for i in range(len(suggestion)):
            gv = gv_pivot.index[suggestion[i]]
            for j in gv:
                if j == gv_name:
                    print(f"You searched '{gv_name}'\n")
                    print("The suggestion gv_tracks are: \n")
                else:
                    print(j)
```

```python
gv_name = "Sleep Music Therapy: Therapy for Sleep Problems, Deep Sleep Music"
recommend_gv_tracks(gv_name)
```
